```python
#Deleting of highly correlating features
numerical_opt = numerical.drop(['Standing Tackle','Sliding Tackle','Marking'], axis=1)
numerical_opt = numerical_opt.drop(['GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes'], axis=1)
display(numerical_opt.head())
```

```python
#Normalizing the data with MinMax Scaler
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler

#Performing X-y split
X = numerical_opt.drop(['OVA'], axis=1)
y = numerical_opt['OVA']


#Normalizing data with MinMaxScaler
transformer = MinMaxScaler().fit(X)
x_normalized = transformer.transform(X)
print(x_normalized.shape)
X_numerical_norm = pd.DataFrame(x_normalized, columns=X.columns)
X_numerical_norm.head()
```

```python
#Just keep the features which have a higher correlation than 0.6 to the target OVA
X_numericals = pd.DataFrame(data=X_numerical_norm, columns=['DRI', 'PAS', 'Base Stats', 'Total Stats', 'Composure', 'Reactions', 'Growth'])
display(X_numericals.head())
```

```python
#Check the correlation between the remaining features
correlations_matrix_X = X_numericals.corr()
display(correlations_matrix_X)
sns.heatmap(correlations_matrix_X, annot=True)
plt.show()
```

```python
#Working with the categoricals
#Transforming the categoricals into multiple numerical columns: 
#OneHot Encoder for categoricals
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(categoricals)
all_columns = encoder.get_feature_names_out(input_features=categoricals.columns)
X_categoricals_encode = pd.DataFrame(encoder.transform(categoricals).toarray(), columns=all_columns)
X_categoricals_encode.head()
display(X_categoricals_encode.shape)
```

```python
def cleandata(data):

    data=data.drop('Loan Date End', axis=1)
    data["Club"]=data["Club"].fillna("Unknown")
    data["Position"]=data["Position"].fillna("Unknown")
    data = data[data['Joined'].isna()==False]
    mean=round(data["Composure"].mean())
    data["Composure"]=data["Composure"].fillna(mean)
    data = data[data['A/W'].isna()==False]
    data.to_csv('./cleaned_data.csv', index_label=False)

    return data.head()
```
